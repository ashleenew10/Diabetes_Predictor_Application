**Undersampling**

```python
import pandas as pd

# Determine the smallest class count
min_count = df['Diabetes_binary'].value_counts().min()

# Undersample each class to min_count
parts = []
for cls, grp in df.groupby('Diabetes_binary'):
    if len(grp) > min_count:
        parts.append(grp.sample(n=min_count, random_state=42))
    else:
        parts.append(grp)

# Recombine
df_balanced = (
    pd.concat(parts)
      .sample(frac=1, random_state=42)
      .reset_index(drop=True)
)

print(df_balanced['Diabetes_binary'].value_counts())
```

#Test Data

```python
#scale and split data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

# split data
X = df_balanced.drop(columns='Diabetes_binary')
y = df_balanced['Diabetes_binary']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    stratify=y,
    test_size=0.2,
    random_state=42
)

# scale on fit only
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)



```

# Random Forest

```python
#importing libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
min_leaf_list = [1, 2, 5, 10, 20, 50, 100]
results = []
```

```python
for n in min_leaf_list:
    rf = RandomForestClassifier(
        n_estimators=300, #this was decided by hypeparameter tuning on this list [50,100,200,300,500]
        max_depth= None, #this was decided by hyperparameter tuning on this list[None,50,100,200,300,500]
        min_samples_leaf=10, #this was decided by hyperparameter tuning on this list [1, 2, 5, 10, 20, 50, 100]
        random_state=42,
        n_jobs=-1
    )
    # Fit
    rf.fit(X_train_scaled, y_train)
    # Predict
    y_pred = rf.predict(X_test_scaled)
    y_proba = rf.predict_proba(X_test_scaled)[:, 1]
    # Compute metrics
    acc   = accuracy_score(y_test, y_pred)
    prec  = precision_score(y_test, y_pred)
    rec   = recall_score(y_test, y_pred)
    f1    = f1_score(y_test, y_pred)
    roc   = roc_auc_score(y_test, y_proba)
    # Store
    results.append({
        "max_depth": n,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1_score": f1,
        "roc_auc": roc
    })
    #print results
    print(f"max_sample_leaf={n} → accuracy={acc:.3f}, precision={prec:.3f}, recall={rec:.3f}, f1={f1:.3f}, roc_auc={roc:.3f}")
```

## The following code creates .pkl file for RANDOM FOREST Model

```python
# Train final Random Forest using best params
final_rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

# Fit on full training data
final_rf.fit(X_train_scaled, y_train)

# Save model to disk
import pickle
with open("diabetes_random_forest_model.pkl", "wb") as f:
    pickle.dump(final_rf, f)
```

# XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
learning_rate_list = [0.01, 0.05, 0.1, 0.2]

results = []
```

```python
for n in learning_rate_list :
    xgb = XGBClassifier(
        objective="binary:logistic",
        eval_metric="logloss",
        use_label_encoder=False,
        n_estimators=50, # #this was decided by hyperparameter tuning on this list [50, 100, 200, 300, 500]
        max_depth=6, # this was decided by hyperparameter tuning on this list  [3, 5, 6, 7, 9, 11, 13, 15]
        learning_rate=n,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )
    # fit on the scaled training data
    xgb.fit(X_train_scaled, y_train)
    # predict
    y_pred = xgb.predict(X_test_scaled)
    y_proba = xgb.predict_proba(X_test_scaled)[:, 1]
    # Compute metrics
    acc   = accuracy_score(y_test, y_pred)
    prec  = precision_score(y_test, y_pred)
    rec   = recall_score(y_test, y_pred)
    f1    = f1_score(y_test, y_pred)
    roc   = roc_auc_score(y_test, y_proba)
    # results
    results.append({
        "learning_rate": n,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1_score": f1,
        "roc_auc": roc
    })
    # 6) summary
    print(f"learning_rate={n} → acc={acc:.3f}, prec={prec:.3f}, "
          f"rec={rec:.3f}, f1={f1:.3f}, roc_auc={roc:.3f}")
```

a learning rate of 0.1 yields the best performance results


```python
# === Train Final XGBoost Model with Best Learning Rate ===
from xgboost import XGBClassifier
import pickle

best_xgb = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    use_label_encoder=False,
    n_estimators=300,        # tuned
    max_depth=6,             # tuned
    learning_rate=0.1,       # best from tuning
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

# Fit on full training data
best_xgb.fit(X_train_scaled, y_train)

# Save to .pkl
with open("diabetes_xgboost_model.pkl", "wb") as f:
    pickle.dump(best_xgb, f)
```

# Logistic Regression

```python
# Dropping insignificant features
insignificant = ["Smoker", "Fruits", "Veggies", "AnyHealthcare","NoDocbcCost"]
df_significant = df_balanced.drop(columns=insignificant)
```

```python
#split into X and y
X = df_significant.drop(columns="Diabetes_binary")
y = df_significant["Diabetes_binary"]

# train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    stratify=y,
    test_size=0.2,
    random_state=42
)
```

```python
# Fit scaler on train
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
#--- Saving a new scaler for logistic regression specifically ---

with open("lr_scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

```python
# predict with Logistic Regression
clf = LogisticRegression(
    solver='liblinear',
    class_weight='balanced',
    random_state=42,
    max_iter=1000
)
clf.fit(X_train_scaled, y_train)

y_pred  = clf.predict(X_test_scaled)
y_proba = clf.predict_proba(X_test_scaled)[:, 1]
```

```python
#metrics and evaluate regression model
print(f"Accuracy:  {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall:    {recall_score(y_test, y_pred):.3f}")
print(f"F1-score:  {f1_score(y_test, y_pred):.3f}")
print(f"ROC AUC:   {roc_auc_score(y_test, y_proba):.3f}\n")

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred), "\n")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import statsmodels.api as sm
#see p-values and coefficients
X_train_df = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)
X_test_df = pd.DataFrame(
    X_test_scaled,
    columns=X_train.columns,
    index=X_test.index
)

# intercept column
X_train_sm = sm.add_constant(X_train_df)

# Fit the logistic regression with statsmodels
logit = sm.Logit(y_train, X_train_sm)
result = logit.fit(disp=False)

# coefficients and p-values
print("Coefficients:\n", result.params, "\n")
print("P-values:\n", result.pvalues, "\n")

# Full summary table
print(result.summary2())
```
